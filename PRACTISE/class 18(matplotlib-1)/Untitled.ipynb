```python
size = 50
df2 = pd.DataFrame({"CNIC": np.random.randint(1,size, size=(size)),
                   "Account No":np.random.randint(1,10, size=(size)),
                   "Mobile No":np.random.randint(2001,3000, size=(size)),
                  "DOB":pd.date_range('1690-01-01', periods=size, freq="M"),
                  "CustomerName":[chr(x) for x in np.random.randint(65,91, size=(size))],
                   "Gender":np.random.choice(["M","F"], size=(size)),
                    "currency":np.random.choice(["036","051","106","105"], size=(size)),
                   "Balance": np.random.randint(1,1000000, size=(size))})
df2.drop_duplicates(inplace=True)
print(len(df2))
df2
```

```python
a = df2.groupby([df2.DOB.dt.year,df2.Gender]).Gender.agg('count').unstack()
a['total'] = a.sum(axis=1)
a.plot(kind='bar')
```

```python
days=[1,2,3,4,5]

reading=[3,6,8,9,9]
playing=[4,5,9,3,4]
watching=[13,7,6,8,12]
teaching=[2,15,20,4,5]

plt.plot([],[], color='c', label='gaming', linewidth=4)
plt.plot([],[], color='m', label='reading', linewidth=4)
plt.plot([],[], color='r', label='watching', linewidth=4)
plt.plot([],[], color='k', label='playing', linewidth=5)
plt.plot([],[], color='g', label='teaching', linewidth=6)

plt.stackplot(days,reading,watching,playing,teaching, colors=['c','m','r','k'])

plt.title("PIAIC Demo")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
