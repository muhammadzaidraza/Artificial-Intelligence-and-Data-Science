```python
import numpy as np
import pandas as pd
cols = []
for i1 in range(26):
    cols.append(chr(i1+65))

for i2 in range(26):
    for j2 in range(26):
        alpha = chr(i2+65) + chr(j2+65)
        cols.append(alpha)

for i3 in range(26):
    for j3 in range(26):
        for k3 in range(26):
            beta = chr(i3+65) + chr(j3+65) + chr(k3+65)
            cols.append(beta)

data = np.full((20,16384) , None)
sheet = pd.DataFrame(data, columns = cols[:16384])
sheet
```
