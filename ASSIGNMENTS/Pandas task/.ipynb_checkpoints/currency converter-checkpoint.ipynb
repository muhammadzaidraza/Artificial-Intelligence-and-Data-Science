```python
import numpy as np
import pandas as pd
import PyCurrency_Converter
```

```python
data_size = 5000
transactions = pd.DataFrame({
    "CNIC" : np.random.randint(1,1000,size=(data_size)),
    "Transaction ID" : np.random.randint(1,2,size=(data_size)),
    "Transaction Date" : pd.date_range("1690-01-01", periods=data_size , freq="M"),
    "Description" : np.random.choice(["TOPUP","IBFT","FT"], size = data_size ),
    "City" : np.random.choice(["Karachi","Islamabad","Lahore","Quetta","Peshawar"], size = data_size ),
    "Amount" : np.random.randint(1,1000000,size=(data_size)),
    "currency" : np.random.choice(["978","036","554","682"], size = (data_size) ),
})
transactions
```

```python
table = pd.read_html('https://www.iban.com/currency-codes')[0]
table
```

```python
#table["Number"].astype("int32")
table["Number"].dtype
```

```python
transactions["currency"].dtype
```

```python
transactions["currency"] = transactions["currency"].astype("float64")
```

```python
transactions
```

```python
df = transactions.merge(table, how='left' , left_on=['currency'], right_on=['Number'] )
df
```

```python
del df["Country"]
del df["Currency"]
del df["Number"]
```

```python
df
```

```python
def converter(amount,code):
    total = PyCurrency_Converter.convert(amount, code, 'PKR')
    return total
```

```python
df["PKR Amount"] = df.apply(lambda x :converter(x["Amount"],x["Code"]),axis=1)
```
